```
import numpy as np
import matplotlib.pyplot as plt
p=0.5 #probabilidade de sair cara
lo=[] #lista que armazena num frequencia de ocorrencias
vsim= [] # armazena numero de simulacoes
nmax=1000 #num maximo de simulacoes
saida=[] #armacena as saidas (cara 1 e coroa  0)
for nsim in np.arange(1, nmax, 10):  # simula de 1 ate nmax
    ncaras= 0 #numero de caras
    for i in range (1,nsim):
       if (np.random.uniform() < p):  # é menor que p (cara)
           ncaras=ncaras+1  #incr contador de caras
    lo.append(ncaras/nsim)       
    vsim.append(nsim)

    #resultados em forma de figura

plt.figure(figsize=(8,6))
plt.plot(vsim, lo, linestyle='-', color="blue", linewidth=2, label="valor simulado")
plt.axhline(y=p, color="r", linestyle='--', label="valor teorico")
plt.ylabel("fracao de caras", fontsize=20)
plt.xlabel("número de experimentos", fontsize=20)
plt.xlim([0.0, nmax])
plt.ylim([0.0, 1.0])
plt.legend()
plt.show(True)



```
